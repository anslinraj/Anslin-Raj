```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
data = pd.read_csv("online.csv")
data
```

```python
data.isna().sum()
```

```python
data = pd.get_dummies(data, columns=['Area'])
data
```

```python
data = data.replace({True: 1, False: 0})
data
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns



correlation_matrix =data.corr()

sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(16, 4))

ax1.scatter(data["Marketing Spend"], data["Profit"])
ax1.set_title("Spend VS Profit")

ax2.scatter(data["Administration"], data["Profit"])
ax2.set_title("Administration VS Profit")

ax3.scatter(data["Transport"], data["Profit"])
ax3.set_title("Transport VS Profit")

plt.tight_layout()
plt.show()
```

```python
x = data["Marketing Spend"]
x
```

```python
y = data["Profit"]
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,train_size=0.7,test_size=0.3,random_state=100)
```

```python
import statsmodels.api as sm
# ADD A CONSTANT 
X_train_sm = sm.add_constant(X_train)
X_train_sm
```

```python
lr = sm.OLS(Y_train,X_train_sm).fit()
lr.params
```
